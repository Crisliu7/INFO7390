```python
import pandas as pd
import numpy as np
import urllib
import requests
import rpy2
import gc
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
loan = ["LoanStats3a","LoanStats3b","LoanStats3c","LoanStats3d",\
        "LoanStats_2016Q1","LoanStats_2016Q2","LoanStats_2016Q3",\
        "LoanStats_2016Q4","LoanStats_2017Q1","LoanStats_2017Q2"]
reject = ["RejectStatsA","RejectStatsB","RejectStatsD","RejectStats_2016Q1",\
          "RejectStats_2016Q2","RejectStats_2016Q3","RejectStats_2016Q4",\
          "RejectStats_2017Q1","RejectStats_2017Q2"]

def download(url_list):
    
    for u in url_list:
        url = ("https://resources.lendingclub.com/%s.csv.zip" % u)

        # Log and print
        operation = "Start retrieving files from " + u + "\n"
        log(operation)
        print(operation)

        # Retrieving
        filename = "/Users/CJL-RMBP/Documents/assignment3data/%s.csv.zip" % u 
        urllib.request.urlretrieve(url,filename=filename)

        # Log and print
        operation = "Done retrieving."
        log(operation)
        print(operation)
    
def log(operation):
    '''Log all the operations with a time stamp into a log file'''

    with open("log_file.txt", 'a') as f:
        data = time.ctime() + ": " + operation
        f.write(data)

download(loan)
download(reject)
```

```python
frames = []
for u in loan:
    
    ts = pd.read_csv("~/Documents/assignment3data/%s.csv"%u,skiprows=1)
    print(ts.shape)
    frames.append(ts)
    del ts
loandata = pd.concat(frames)
gc.collect()
```

```python
print(loandata.shape)
loandata.head()
```

```python
loandata.describe()
```

```python
loan_mv = loandata.isnull().sum(axis=0).reset_index()
loan_mv.columns = ['column_name', 'no_missing']
loan_mv['missing_ratio'] = loan_mv['no_missing'] / loandata.shape[0]
loan_mv.sort_values(by='no_missing',ascending=True, inplace=True)
loan_mv
```

```python
co = loan_mv[loan_mv['missing_ratio']>0.7]['column_name'].tolist()
loandata = loandata.drop(co,axis=1)
```

```python
print(loandata.shape)
loandata.head()
```

```python
loandata.to_csv("./loandata.csv",index=None)
```
