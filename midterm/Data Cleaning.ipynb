```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn.ensemble import RandomForestRegressor
from sklearn import decomposition
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Load the training set
ts = pd.read_csv('./Data/training_set_dirty.csv')
print(ts.shape)
ts.head()
```

```python
ts['longitude'] = ts['longitude']/1000000
ts['latitude'] = ts['latitude']/1000000
ts.head()
```

```python
# For date
def transac(df):
    df['transactiondate']=pd.to_datetime(df['transactiondate'])
    df['month'] = df['transactiondate'].dt.month
    df.drop('transactiondate', axis=1, inplace=True)
transac(ts)
ts.head()
```

```python
# Check the missing values
ts_mv = ts.isnull().sum(axis=0).reset_index()
ts_mv.columns = ['column_name', 'no_missing']
ts_mv['missing_ratio'] = ts_mv['no_missing'] / ts.shape[0]
ts_mv.sort_values(by='no_missing',ascending=True, inplace=True)
ts_mv
```

```python
# Columns to be dropped 
# because of high data missing rate(>0.7)
drop = ts_mv[ts_mv['missing_ratio']>0.7]['column_name'].tolist()
print(drop)
```

```python
for a in drop:
    ts.drop(a, axis=1, inplace=True)
ts.drop('year_y', axis=1, inplace=True)
print(ts.shape)
ts.head()
```

```python
# Check all empty rows with no feature values
temp_a = ts.set_index(['parcelid','logerror','year_x','month'])
print(temp_a.shape)
temp_b = temp_a.dropna(axis=0, how='all')
print(temp_b.shape)
print("There are %s lines in the data set that have no feature values" % (temp_a.shape[0]-temp_b.shape[0]))
```

```python
# Drop the 119 lines
ts = temp_b.reset_index()
print(ts.shape)
ts.head()
```

```python
ts.dtypes
```

```python
# For missing categorical data, fill with mode
print(ts['regionidzip'].mode())
print(ts['regionidcity'].mode())
print(ts['regionidneighborhood'].mode())

print(ts['propertycountylandusecode'].mode())
print(ts['propertyzoningdesc'].mode())

print(ts['yearbuilt'].mode())

print(ts['buildingqualitytypeid'].mode())
print(ts['heatingorsystemtypeid'].mode())
print(ts['airconditioningtypeid'].mode())
```

```python
# For missing categorical data, fill with mode
ts['regionidzip'].fillna('97319', inplace=True)
ts['regionidcity'].fillna('12447', inplace=True)
ts['regionidneighborhood'].fillna('27080', inplace=True)

ts['propertycountylandusecode'].fillna('0100', inplace=True)
ts['propertyzoningdesc'].fillna('LAR1', inplace=True)

ts['yearbuilt'].fillna(1955, inplace=True)

ts['buildingqualitytypeid'].fillna('7', inplace=True)
ts['heatingorsystemtypeid'].fillna('2', inplace=True)
ts['airconditioningtypeid'].fillna('1', inplace=True)
```

```python
# For missing numerical data, fill with mean
mean_values = ts.mean(axis=0)
ts = ts.fillna(mean_values, inplace=True)
```

```python
ts_mv = ts.isnull().sum(axis=0).reset_index()
ts_mv.columns = ['column_name', 'no_missing']
ts_mv['missing_ratio'] = ts_mv['no_missing'] / ts.shape[0]
ts_mv.sort_values(by='no_missing',ascending=True, inplace=True)
ts_mv
################### Missing values filled ###################
```

```python
################### Cleaning 'logerror' ###################
f, ax = plt.subplots(figsize=(6,3))
ax.set_title("Box of logerror")
color = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')
ts['logerror'].plot.box(color=color, sym='r+')
plt.show()
# Quantiles
ts['logerror'].quantile([.00001, .001, .1, .25, .5, .75, .9, .999, .99999])
```

```python
# Outliers
co = ts['logerror'].loc[(ts['logerror']>0.2) | (ts['logerror'] < -0.3)]
# Replace
ts['logerror'] = ts[['logerror']].replace(co.values, ts['logerror'].mean())
```

```python
co = ts['logerror'].loc[(ts['logerror']>0.2) | (ts['logerror'] < -0.3)]
print(co)
f, ax = plt.subplots(figsize=(6,3))
ax.set_title("Box of logerror - cleaned")
color = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')
ts['logerror'].plot.box(color=color, sym='r+')
plt.show()
################### Cleaned for 'logerror' ###################
```
